### Observations:

- *Call Option – Spot Price Sensitivity:*  
- *Put Option – Time Sensitivity:*  
- *Put Option – Spot Price Sensitivity:*  

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import seaborn as sns
sns.set_style('darkgrid')

def bs_call(S0, K, sigma, t=1, r=0):
    """
    Description:

    Computes the Black-Scholes value of a European call option.

    Parameters:
        S0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        t: Time to expiration (in years)
        r: Risk-free interest rate

    Returns:
        Call option price
    """
    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    #d2 = (np.log(S0/K) + (r-.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    return S0*norm.cdf(d1)-K*np.exp(-r*t)*norm.cdf(d2)


def bs_put(S0, K, sigma, t=1, r=0):
    """
    Description:

    Computes the Black-Scholes value of a European put option.

    Parameters:
        S0: Current asset price
        K: Strike price
        sigma: Yearly standard deviation of log-returns (volatility)
        t: Time to expiration (in years)
        r: Risk-free interest rate

    Returns:
        Put option price
    """

    d1 = (np.log(S0/K) + (r+.5*sigma**2)*t)/(sigma*np.sqrt(t))
    #d2 = (np.log(S0/K) + (r-.5*sigma**2)*t)/(sigma*np.sqrt(t))
    d2 = d1 - sigma*np.sqrt(t)
    return -S0*norm.cdf(-d1)+K*np.exp(-r*t)*norm.cdf(-d2)


S0 = 100
K = 110
sigma = .3
times = np.linspace(1/12, 5, 70)


call_prices = [bs_call(S0,K,sigma, t = times[i]) for i in range(len(times))]


plt.figure(figsize = (12,8))
plt.title('Changes Black-Scholes Call Option prices time to expiration', size = 25)
plt.plot(times, call_prices)
plt.xlabel('Time to Expiration')
plt.ylabel('Black-Scholes Call Price')
plt.show()
```

#### (B) Create a separate graphic to visualize how the *rate of change* of the Black-Scholes call option price depends on the spot price $S(0)$.

```python
S0 = np.linspace(100, 120, 100)
K = 115
sigma = .3
time = 1

call_prices = [bs_call(S0[i],K,sigma, time) for i in range(len(S0))]


plt.figure(figsize = (12,8))
plt.title('Changes Black-Scholes Call Option prices vs Spot Price', size = 25)
plt.plot(S0, call_prices)
plt.xlabel('Spot Price')
plt.ylabel('Black-Scholes Call Price')
plt.show()
```

### (C) Repeat parts (A) and (B) for **put option** prices.

```python
S0 = 100
K = 110
sigma = .3
times = np.linspace(1/12, 5, 70)


put_prices = [bs_put(S0,K,sigma, t = times[i]) for i in range(len(times))]


plt.figure(figsize = (12,8))
plt.title('Changes Black-Scholes Put Option prices time to expiration', size = 25)
plt.plot(times, put_prices)
plt.xlabel('Time to Expiration')
plt.ylabel('Black-Scholes Put Price')
plt.show()
```

```python
S0 = np.linspace(100, 120, 100)
K = 115
sigma = .3
time = 1

put_prices = [bs_put(S0[i],K,sigma, time) for i in range(len(S0))]


plt.figure(figsize = (12,8))
plt.title('Changes Black-Scholes Put Option prices vs Spot Price', size = 25)
plt.plot(S0, put_prices)
plt.xlabel('Spot Price')
plt.ylabel('Black-Scholes Call Price')
plt.show()
```

The rate of change of the Black-Scholes call option price behaves as time progresses is called Theta, representated by $\Theta$. As the time progress and as we move towards the time of expiration, the rate of change of Black-Scholes call option price is negative and it become more and more negative with time. The reason for this is that as the time to expiration approaches, the time required for the price of underlying asset to move in the favourable direction decreases. This decrease in the opportunity means that the option losses its value. From the seller point of view, the negatixve $\Theta$ is good as the time passes, the option will lose its value with time and the seller can buy the option back cheapter or the option will become worthless.

The rate of change of the Black-Scholes call option price with respect the spot price is known as Delta, represented by $\Delta$. The delta of a call option increases as the spot prices increases. The $\Delta$ lies between $0$ and $1$. For each unit increase in spot price, call price changes by $\Delta$. If the spot price is much lower than the strike price then the $\Delta$ will be close to $0$. On the other hand, if the spot price is close to the strike price then the $\Delta$ will be closer to $1$.
